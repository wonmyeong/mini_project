```python
import pandas as pd
```

```python
df_women_1 = pd.read_csv("women_rating.csv")
df_women_1
```

```python
df1=pd.read_csv("musinsa_women.csv")
df1
```

```python
# 인덱스를 기준으로 두 데이터프레임 합치기
merged_df = pd.merge(df1, df_women_1, how='outer', left_index=True, right_index=True)


merged_df.rename(columns={'평점': 'rating'}, inplace=True)
merged_df=merged_df.drop(columns=['Item'])
merged_df['rating'] = merged_df['rating'].round(1)
```

```python
merged_df
```

```python
merged_df.to_csv('final_musinsa_women.csv', index=False,encoding='utf-8-sig') 
```
